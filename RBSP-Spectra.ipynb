```python
#import spacepy and cdasws
import spacepy
from spacepy.coordinates import Coords
from spacepy.time import Ticktock
import spacepy.datamodel as dm
import spacepy.plot as splot
from cdasws import CdasWs

#import standard python stuff
import json
import datetime

#import numpy and matplotlib
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.gridspec as gridspec

#Kaipy and related
import kaipy.kaiViz as kv
import kaipy.kaiTools as kaiTools
```

```python
t0 = datetime.datetime(2012, 10, 25, 8, 0, 0)
t1 = datetime.datetime(2012, 10, 25, 10, 0,0)
```

```python
cdas = CdasWs()
```

```python
status,data = cdas.get_data('RBSPA_REL03_ECT-REPT-SCI-L2',['FESA'],t0,t1)
```

```python
cmapName = 'viridis'
vMin=1
vMax=1.0e6
norm = kv.genNorm(vMin,vMax,doLog=True)
fig = plt.figure(figsize=(10,5))
gs = gridspec.GridSpec(1,2,width_ratios=[10,0.25],wspace=0.025)
Ax = fig.add_subplot(gs[0,0])
AxC = fig.add_subplot(gs[0,1])
#Not great to manipulate data, but lognorm and pcolormesh cannot handle zero value
data['FESA'][data['FESA']==0]=1.0e-5
Ax.pcolormesh(data['Epoch'],np.asarray(data['FESA_Energy']),
                          np.transpose(np.asarray(data['FESA'])),
               norm=norm,cmap=cmapName)
Ax.set_yscale('log')
kv.genCB(AxC,norm,cM=cmapName,doVert=True)
kv.SetAxDate(Ax)
```
